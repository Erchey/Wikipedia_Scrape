```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import csv
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_largest_universities')
```

```python
soup = BeautifulSoup(response.content)
soup
```

```python
table = soup.find("table", attrs={"class": "sortable"})
```

```python
trs = table.findAll('tr')
trs
```

```python
def extract_rows(tr):
    rows = []
    rows.append(list(map(lambda yu: yu.text.strip(), tr[0].findAll('th'))))
    for row in tr[1:]:
        single_row = list(map(lambda yu: yu.text.strip(), row.findAll('td')))
        if len(single_row) < 6:
            single_row.append("")
        link = row.findAll('td')[1].a.attrs["href"]
        single_row[-1] = f'https://en.wikipedia.org{link}' if link else ""
        rows.append(single_row)

    rows[0][-1] = 'Link'
    with open('universities.csv', 'w', newline="") as f:
        csvwriter = csv.writer(f)
        csvwriter.writerows(rows)
    df = pd.DataFrame(rows[1:], columns=rows[0])
    return df.head(30)

extract_rows(trs)
```
